```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('company_sales_data.csv')
```

```python
sales_data
```

```python
plt.plot(sales_data['month_number'], sales_data['total_profit'])
plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize = (7, 6))
plt.plot(sales_data['month_number'], sales_data['facecream'], label = 'Face Cream')
plt.plot(sales_data['month_number'], sales_data['facewash'], label = 'Facewash')
plt.plot(sales_data['month_number'], sales_data['toothpaste'], label = 'Toothpaste')
plt.plot(sales_data['month_number'], sales_data['bathingsoap'], label = 'Bathing Soap')
plt.plot(sales_data['month_number'], sales_data['shampoo'], label = 'Shampoo')
plt.plot(sales_data['month_number'], sales_data['moisturizer'], label = 'Moisturizer')
plt.title('Sales of Different Commodities by Month')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
total_sales = []
for column in sales_data.columns[1:7]:
    total_sales.append(sales_data[column].sum())
```

```python
plt.pie(total_sales, labels = ['Face Cream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.title('Total Sales over Last Year')
plt.show()
```
